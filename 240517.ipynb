```python
iris_data=pd.read_csv('iris.csv')
sns.scatterplot(x='petal_length',y='petal_width'
,hue='species',data= iris_data)
plt.title('꽃 그래프', fontsize=20)
plt.show()
```

```python
x1 = np.random.uniform(0, 10, 100)
y1 = np.sin(x1)

x2 = np.random.uniform(0, 10, 100)
y2 = np.cos(x2)

# 1x2 그리드 형태의 서브플롯 생성
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# 첫 번째 서브플롯
ax1.scatter(x1, y1, color='blue', label='sin그래프')
ax1.set_title('Data Set 1: sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()

# 두 번째 서브플롯
ax2.scatter(x2, y2, color='red', label='cos그래프')
ax2.set_title('Data Set 2: cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()

plt.show()
```
